```python
df['Lead_Creation_Date'] = pd.to_datetime(df['Lead_Creation_Date'], format = '%m%d%Y')
df['DOB'] = pd.to_datetime(df['DOB'], format = '%d/%m/%y')
```

```python
df.info()
```

```python
df['Approved'].unique()
```

```python
leads_created = df['Lead_Creation_Date'].count()
```

```python
contacted_leads = df[df['Contacted'] == 'Y']['Contacted'].count()
```

```python
approved_leads = df[df['Approved'] == 1]['Approved'].count()
```

```python
print(leads_created, contacted_leads, approved_leads)
```

```python
df['Source_Category'].unique()
```

```python
funnel_data = pd.DataFrame({
    'Funnel Stage': ['Leads Created', 'Contacted Leads', 'Approved Leads'],
    'Count': [leads_created, contacted_leads, approved_leads]
})
```

```python
funnel_data
```

```python
funnel_data.to_csv('funnel_data.csv', index = False)
```
